# Data Collection

```python
import os 
import pandas as pd 
import numpy as npb
```

```python
directory = './SkeletonData/Simplified/'
position_dict = {'stand.txt': 0, 'Stand-frame.txt': 1, 'sit.txt': 2 ,'chair.txt': 3, 'wheelchair.txt': 4}
headers = []
for i in range(25):
    headers.append(f'x{i}')
    headers.append(f'y{i}')
    headers.append(f'z{i}')
headers.append('position')
headers.append('gestureLabel')

for filename in os.listdir(directory):
    print(filename)
    if filename[0] == '.':
        continue
    
    filename_splits = filename.split('_')
    isCorrectGesture = filename_splits[-2]
    gestureLabel = filename_splits[2]
    position = filename_splits[-1]
    
    if isCorrectGesture == '1':
        fileData = pd.read_csv(os.path.join(directory,filename), delimiter=',')
        fileData['position'] = [position_dict[position]] * len(fileData)
        fileData['gestureLabel'] = [gestureLabel] * len(fileData)
        if os.path.exists(f'finalDataset_{gestureLabel}.csv'):
            fileData.to_csv(f'finalDataset_{gestureLabel}.csv', mode='a', index=False, header=False)
        else:
            fileData.to_csv(f'finalDataset_{gestureLabel}.csv', mode='a', index=False, header=headers)
        # fileData.to_csv('finalDataset.csv', mode='a', index=False, header=headers)
        
```
